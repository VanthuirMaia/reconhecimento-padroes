```python
import sys, os, joblib
import numpy as np

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from src.pca_analysis import apply_pca

# Carregar dados escalados
X_train_scaled, X_test_scaled, y_train, y_test = joblib.load('../data/processed/scaled_data.pkl')

# Aplicar PCA
X_train_pca, X_test_pca, pca = apply_pca(X_train_scaled, X_test_scaled)

print("Dimensões após PCA:", X_train_pca.shape)

# Variância explicada individual
print("Variância explicada (individual):")
print(pca.explained_variance_ratio_)

# Variância explicada acumulada
cumulative = np.cumsum(pca.explained_variance_ratio_)
print("\nVariância explicada acumulada:")
print(cumulative)

# Identificar o número mínimo de componentes para atingir ~90%
k = np.argmax(cumulative >= 0.90) + 1
print(f"\nNúmero de componentes necessários para atingir 90% de variância: {k}")
```
